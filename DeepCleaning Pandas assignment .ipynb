```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
df
```

```python
"""Some values in the the FlightNumber column are missing. These numbers are
meant to increase by 10 with each row so 10055 and 10075 need to be put in
place. Fill in these missing numbers and make the column an integer column
(instead of a float column)."""

print(df)
for i in range(df.FlightNumber.count()+1): # taking lenght of the particular columns so that loop can run untill
    if pd.isna(df.FlightNumber.loc[i,]): # appying if condition  
        df.loc[i,"FlightNumber"] = df.FlightNumber.loc[i-1,] +10 # running loop till range -1 and adding 10 
print("*"*50)
df.FlightNumber = df.FlightNumber.interpolate().astype("int64")

print("*"*50)
df
```

```python
"""2. The From_To column would be better as two separate columns! Split each
string on the underscore delimiter _ to give a new temporary DataFrame with
the correct values. Assign the correct column names to this temporary
DataFrame."""

df_temp = df.copy() # creating copy of the DF

df_temp[["From","To"]] = df_temp["From_To"].str.split("_",expand = True) # spliting value with spacified delimeter and using expand parameter to expand it in two columns
df_temp
```

```python
"""3. Notice how the capitalisation of the city names is all mixed up in this
temporary DataFrame. Standardise the strings so that only the first letter is
uppercase (e.g. "londON" should become "London".)"""

def capital(x):
    x = x.str.capitalize()
    return x
df_temp["From_To"] = capital(df_temp["From_To"]) # using capilatization function to get the result. str.capitalize is series fucntion
df_temp["From"] = capital(df_temp["From"])
df_temp["To"] = capital(df_temp["To"])
df_temp
```

```python
"""4. Delete the From_To column from df and attach the temporary DataFrame
from the previous questions."""
df.drop(columns="From_To",inplace =True) # delete a particular coluum 
df["From_To"] = df_temp["From_To"] #replacing a column from one df to another df
df
```

```python
"""5. In the RecentDelays column, the values have been entered into the
DataFrame as a list. We would like each first value in its own column, each
second value in its own column, and so on. If there isn't an Nth value, the value
should be NaN."""


row_list = [] # extract entire dataset into row list and saving it again in the DF. runing loop on recetdelay so that each value seprates 

f = df.apply(lambda x : [row_list.append([x["FlightNumber"],x["Airline"],rd , x["From_To"]]) for rd in x.RecentDelays],axis = 1)

df1 = pd.DataFrame(row_list,columns= df.columns)
df1
```

```python
"""Expand the Series of lists into a DataFrame named delays, rename the columns
delay_1, delay_2, etc. and replace the unwanted RecentDelays column in df
with delays."""

df3 = pd.DataFrame(df.RecentDelays.values.tolist()) # taking out recentdelay into different column using tolist fucntion 
df3
```

```python
lenght = df3.shape[1] # taking all columns ans saving in a variable
col_name = {}  # creating an empty dictionery 
for i in range(lenght): # running a loop and taking out key and defining values for creating the columns 
    key = df3.columns[i]
    value = "Delay" + str(i+1)
    col_name[key] = value
```

```python
df3.rename(columns=col_name, inplace= True)
df3
```
